# Tips Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Set Seaborn as our default plotter

```python
sns.set()
```

## Loading exercise data

```python
data = sns.load_dataset('tips')
data.head(10)
```

## Plot tips rate depending on days and gender

```python
sns.set_theme(style="darkgrid")
g=sns.displot(
    data ,x="tip", col = "day"  , row="sex",binwidth=.5, height=5, facet_kws=dict(margin_titles=True))
g.map(sns.histplot, "tip")
```

## Conclusion :
As the data are showing us we can say that male are more likely to give tips more than female but also we notice that both genders gives less tips on friday.
